```
!pip3 -q install ColabTurtle    # https://github.com/tolgaatam/ColabTurtle
from ColabTurtle.Turtle import *
import random
import math
```

```
def cleanSlate(startX, startY):
    initializeTurtle(initial_window_size=(900,900))
    clear()
    bgcolor('white')
    color('black')
    penup()
    goto(startX, startY)
    #pendown()
```

```
def ShapeShifter(xCentre,yCentre,meanRadius,steps = 12, c = 'black', perturb = 0, freq = 1, phase = 0):
    # permitted colours https://www.w3schools.com/colors/colors_names.asp
    st()
    color(c)
    up()
    #goto(xc+r*math.cos(math.radians(o)),yc+r*math.sin(math.radians(o)))
    #down()
    theta = 0
    delta = 360/steps
    #print('---', e,k,o)
    for step in range(steps+1):
        theta = math.radians(step*delta+phase)
        #print(thetaR)
        radial = meanRadius+perturb*math.sin(freq*math.radians(step*delta))
        #print(current, thetaR, re,xc+re*math.cos(thetaR),yc+re*math.sin(thetaR))
        goto(xCentre+radial*math.cos(theta),yCentre+radial*math.sin(theta))
        down()
    ht()
```

```
cleanSlate(450,400)
cx, cy, radius = 450,450,100
ShapeShifter(cx,cy,radius,24,'orangered')
ShapeShifter(cx,cy,radius,48,'blue',20,3)
ShapeShifter(cx,cy,radius,48,'green',20,3,30)
ShapeShifter(cx,cy,radius,48,'orange',20,3,60)
ShapeShifter(cx,cy,radius,48,'darkmagenta',20,3,90)

cx, cy, radius = 450,450, 150
ShapeShifter(cx,cy,radius,24,'orangered')
#ctcircle(cx,cy,100,60,'GREEN',20,4)
ShapeShifter(cx,cy,radius,120,'crimson',20,12)
ShapeShifter(cx,cy,radius,120,'forestgreen',20,12,10)
ShapeShifter(cx,cy,radius,120,'indigo',20,12,20)
```

```
def sqSpiral(sx,sy,steps,c='red',stride = 100, turn=90):
    st()
    up()
    goto(sx,sy)
    color(c)
    down()
    pensize(1)
    for step in range(steps+4):
        stride = stride+2
        turn = turn+1
        fd(stride)
        rt(turn)
    ht()
```

```
#import turtle
cleanSlate(450,450)
#bgcolor('black')
#colormode(255)
speed(10)
sqSpiral(225,225,60)
```
